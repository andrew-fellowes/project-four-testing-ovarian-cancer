```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Create DataFrame
data_df = pd.read_csv('../Resources/HRD_Agreement.csv')
data_df.dtypes
```

```python
# Replace '-' with NaN in 'SOPHiAGIIndex' column
data_df['SOPHiAGIIndex'] = data_df['SOPHiAGIIndex'].replace('-', np.nan)
```

```python
# Sort the DataFrame by SOPHiAGIIndex
data_df = data_df.sort_values(by='MyriadGIScore')
```

```python
# Define threshold values
myriad_threshold = 42
sophia_threshold = 0
```

```python
# Create scatterplot with seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_df, x='MyriadGIScore', y='SOPHiAGIIndex', hue='Agreement1')

# Plot threshold lines
plt.axhline(y=0, color='r', linestyle='--', label='Sophia Threshold')
plt.axvline(x=42, color='g', linestyle='--', label='Myriad Threshold')

plt.title('Scatterplot of SOPHiAGIIndex vs MyriadGIScore')
plt.xlabel('MyriadGIScore')
plt.ylabel('SOPHiAGIIndex')
plt.xlim(0, 100)  # Set x-axis limits
plt.ylim(-20, 20)  # Set y-axis limits

# Manually set y-axis ticks to display both negative and positive values
plt.yticks(range(-20, 21, 5))

plt.legend()
plt.grid(True)
plt.show()
```
